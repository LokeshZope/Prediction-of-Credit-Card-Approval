## Introduction

   Credit cards are a widely utilised financial instrument. However, because of a number of variables, including employment position and income level, getting a credit card can be challenging for many people. Credit card approval can be predicted by machine learning models, which can assist  the application process and increase the approval rate. 
   The goal of the project is to  investigate the relationship between several parameters and credit card acceptance rates and  apply predictive models to increase the precision of credit card approval forecasts.

## Attributes Description

Ind_id: This column indicates the indivisual id.

GENDER: This column indicates the gender of the individual, either male or female. 
        Type: Categorical (binary).

Car_Owner: This column indicates whether the individual owns a car or not, either Yes or No.
        Type: Categorical (binary).

Propert_Owner: This column indicates whether the individual owns a property or not, either Yes or No.
        Type: Categorical (binary).

CHILDREN: This column indicates the number of children the individual has.
        Type: Numerical(integer).

Annual_income: This column contains the annual income of the individual.
        Type: Numerical (continuous).

Type_Income: This column indicates the type of income the individual earns, such as salary or self-employed income.
        Type: Numerical (nominal).

EDUCATION: This column indicates the level of education of the individual.
        Type: Categorical (ordinal).

Marital_status: This column indicates the marital status of the individual, such as Single, Married, Divorced, etc.
        Type: Categorical (nominal).

Housing_type: This column indicates the type of housing the individual lives in, such as a house or apartment.
        Type: Categorical (nominal).

Birthday_count: This column contains the age of the individual.
        Type: Numerical(integer).

Employed_days: This column indicates the number of days the individual has been employed.
        Type: Numerical(integer).

Mobile_phone, Work_Phone, Phone: These columns indicates the contact information of individual.
        Type: Categorical (binary)
        
EMAIL_ID:These column indicates whether individual has email_id or not
        Type: Categorical (nominal).

Type_Occupation: This column indicates the type of occupation of the individual, such as healthcare or education.
        Type: Categorical (nominal).

Family_Members: This column indicates the number of family members the individual has.
        Type: Numeric (integer).

Label: This column contains the label for credit card approval, either approved or not approved.
        Type: Categorical (binary)

## Splitting the data (Test_Train split)

```python
X = data.drop(columns = 'label')
Y = data['label']
```

```python
#splitting data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
# count the number rows and columns for x_test
X_test.shape
```

```python
# count the number rows and columns for x_train
X_train.shape
```

```python
# count the number rows for y_test
Y_test.shape
```

```python
# count the number rows for y_train
Y_train.shape
```

## Scaling

```python
# importing the MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.fit_transform(X_test)
print(x_train.shape)
print(x_test.shape)
```

## Logistic Regression Classifier

```python
# importing the LogisticRegression
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train, Y_train)
```

```python
#Importing scikit-learn metrics module for accuracy calculation
from sklearn import metrics
from sklearn.metrics import accuracy_score
y_pred = log_reg.predict(x_test)
print("logistic regression training set:",log_reg.score(X_train, Y_train))
print("logistic regression testing set:",log_reg.score(X_test, Y_test))
print("Accuracy of logistic regression classifier: ", accuracy_score(Y_test,y_pred))
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit
model = LogisticRegression()
ssplit =ShuffleSplit(n_splits=10,test_size=0.20)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,Y,cv=ssplit)
print(results)
print(np.mean(results))
```

## Decision tree classifier

```python
# decision tree classification 
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier()
dec_tree.fit(X_train,Y_train)
```

```python
y_pred = dec_tree.predict(x_test)
print("decision tree training set:",dec_tree.score(X_train, Y_train))
print("decision tree testing set:",dec_tree.score(X_test, Y_test))
print("Accuracy of decision tree classifier :", accuracy_score(Y_test,y_pred))
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit
model= DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.20)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,Y,cv=ssplit)
print(results)
print(np.mean(results))
```

## Random Forest Classifier

```python
#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
y_pred = rf.predict(X_test)
print("rf training set:",rf.score(X_train, Y_train))
print("rf testing set:",rf.score(X_test, Y_test))
print("Random Forest classifier has accuracy of: ", accuracy_score(Y_test, y_pred))
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit
model=RandomForestClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.20)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,Y,cv=ssplit)
print(results)
print(np.mean(results))
```

## KNN Classifier
